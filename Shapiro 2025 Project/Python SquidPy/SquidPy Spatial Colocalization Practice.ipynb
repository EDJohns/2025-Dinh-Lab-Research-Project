#### Spatial Graphs

The spatial graph identifies neighboors to individual cells. By default there is not a distance weight but rather cells are treated as neighboors or not neighboors (1 or 0). Typically only cells in direct contact would be considered neighboors but different types of datasets have peculiar behaviors which may negativly impact functionallity. 

Setup of the function:
* coord_type: For these datasets coord_type="generic" should be used used. Assumes non structured 2D coordinates not regular grid. Only visium data uses a regulaar grid.
* Neighboor Definition: This specifies how neighboors are defined in a particular case.
  - n_neigh=<int> instead of distance it define neighboors as the closest <int> cells.
  - radius=<float> cells within custom radius are considered neighboors
  - delaunay=True uses the delauny triagulation. Creates triangles to "naturally" connect spatial neighboors without having to input radius or a set number of cases. Good for infering geometry from evenly distributed non-grid. Probably the best option for my use case.

Data Outputs:
* adata.obsp[‘spatial_connectivities’]: adjacency and weighted adjacency matricies are saved here
* adata.obsp[‘spatial_distances’]: sparse adjacency matrix

#### Neighborhood Enrichment Test

More straightforward than the spatial graphs. Basically looks at the interaction between 2 cell types compared to random configuration. Outputs a Z-score of the results 

Data Outputs:
* adata.uns["nhood_enrichment"]["z"]
- Each entry is a z-score of how often one cell group of another cell type.

## Function to output Z maps for a given protein frame

```python
def Creat_Z_map(adata, sample_string):
    """
    Retrieve the expression value, cell metadata, and gene metadata from an AnnData object.

    Parameters:
    - adata: AnnData object
    - cell_name: str, name of the cell (must be in adata.obs_names)
    - gene_name: str, name of the gene (must be in adata.var_names)

    Returns:
    - dict with expression value, cell metadata (pandas Series), and gene metadata (pandas Series)
    """
    # Makes subset of the adata object
    adata_sub = adata[adata.obs.cell_ID.str.contains(sample_string)].copy()
    
    #Compute spatial matrix 
    sq.gr.spatial_neighbors(adata_sub, coord_type="generic",delaunay=True)
    
    # Creates Matrix to Display
    dense_df = pd.DataFrame(
    adata_sub.obsp["spatial_connectivities"].toarray(),
    index=adata_sub.obs_names,
    columns=adata_sub.obs_names
    )

    # Ensure 'cell type' is categorical 
    adata_sub.obs['merged_annot_cluster'] = adata_sub.obs['merged_annot_cluster'].astype('category')

    # Run neighborhood enrichment on the subset
    sq.gr.nhood_enrichment(adata_sub, cluster_key="merged_annot_cluster")
    
    # Extract Z-scores and labels
    adata_sub.uns
    z = adata_sub.uns["merged_annot_cluster_nhood_enrichment"]["zscore"]
    RawNums = adata_sub.uns["merged_annot_cluster_nhood_enrichment"]["count"]
    
    # Convert to labeled DataFrame
    z_df = pd.DataFrame(z)
    RawNums_df = pd.DataFrame(RawNums)
    
    
    # Get the Z-score matrix
    z = adata_sub.uns["merged_annot_cluster_nhood_enrichment"]["zscore"]
    
    # Manually get cluster names from adata_sub.obs["merged_annot_cluster"]
    cluster_labels = np.unique(adata_sub.obs["merged_annot_cluster"])
    
    # Sanity check that dimensions match
    assert z.shape == (len(cluster_labels), len(cluster_labels)), "Mismatch in shape"
    
    # Make labeled matrix
    z_df = pd.DataFrame(z, index=cluster_labels, columns=cluster_labels)
    RawNums_df = pd.DataFrame(RawNums, index=cluster_labels, columns=cluster_labels)
    
    # Display
    display(RawNums_df)
    display(z_df.round(2))

    plt.figure(figsize=(10, 8))
    sns.heatmap(
        z_df,
        cmap="coolwarm",     # diverging color map: red = enriched, blue = depleted
        center=0,            # center at 0 for symmetric color scale
        annot=True,          # show numbers in each cell
        fmt=".1f",           # format the Z-scores
        linewidths=0.5,      # grid lines
        cbar_kws={"label": "Z-score"}
    )
    plt.title("Neighborhood Enrichment Z-score Heatmap")
    plt.xlabel("Neighbor Cell Type")
    plt.ylabel("Anchor Cell Type")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()

    return {
        "Cell Spatial Matrix": dense_df
    }


result = Creat_Z_map(adata, "c_1_3")
display(result)
```
